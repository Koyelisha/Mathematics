```python
from PIL import Image
from matplotlib import pyplot as plt
import numpy as np
```

```python
! wget https://raw.githubusercontent.com/jonkrohn/DLTFpT/master/notebooks/oboe-with-book.jpg
```

```python
img = Image.open('oboe-with-book.jpg')
_ = plt.imshow(img)
```

```python
imggray = img.convert('LA')
plt.imshow(imggray)
```

```python
img = np.array(list(imggray.getdata(band=0)),float)
img.shape=(imggray.size[1],imggray.size[0])
imgmat = np.matrix(img)
plt.imshow(imgmat,cmap='gray')
```

```python
u,d,vt = np.linalg.svd(imgmat)
```

```python
reconsisting = np.matrix(u[:,:1]) * np.diag(d[:1]) * np.matrix(vt[:1,:])
plt.imshow(reconsisting,cmap='gray')
```

```python
for i in [2,4,8,16,32,64,128]:
  reconsisting = np.matrix(u[:,:i]) * np.diag(d[:i]) * np.matrix(vt[:i,:])
  plt.imshow(reconsisting,cmap='gray')
  title = "n= %s" % i
  plt.title(title)
  plt.show()
```
