```python
from matplotlib import pyplot as plt
import numpy as np
import tensorflow as tf
```

```python
def plot_vectors(vectors,colors):
    plt.figure()
    plt.axhline(y=0,color='lightgray')
    plt.axvline(x=0,color='lightgray')
    for i in range(len(vectors)):
        x = np.concatenate([[0,0],vectors[i]])
        plt.quiver([x[0]],[x[1]],[x[2]],[x[3]],angles='xy',scale_units='xy',scale=1,color=colors[i])
        # plt.show()
```

```python
v = np.array([3,1])
v
```

```python
plot_vectors([v],['lightblue'])
```

```python
I = np.array([[1,0],[0,1]])
I
```

```python
VI = np.dot(v,I)
VI
```

```python
#eigen decomposition
#A = v^vinv
A = np.array([[2,3],[6,7]])
A
```

```python
lambda1,v1 = np.linalg.eig(A)
v1
```

```python
lamb1 = np.diag(lambda1)
lamb1
```

```python
v1inv = np.linalg.inv(v1)
v1inv
```

```python
vupl = np.dot(v1,lamb1)
vupl
```

```python
v = np.dot(vupl,v1inv)
v
```
