y = x^2 + 2x + 2

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def f(my_x):
    my_y = my_x**2 + 2*my_x + 2
    return my_y
```

```python
x = np.linspace(-10,10,10000)
```

```python
y = f(x)
```

```python
fig, ax = plt.subplots()
plt.axhline(y=0,color='lightgray')
plt.axvline(x=0,color='lightgray')
plt.plot(x,y)
```

let's identify the slope where x = 2

```python
f(2)
```

call the point p which located at at (2,10)

```python
fig,ax = plt.subplots()
plt.axhline(y=0,color='lightgray')
plt.axvline(x=0,color='lightgray')
plt.plot(x,y)
plt.scatter(2,10,color='red',zorder=2)
```

The delta method uses the difference b/w two points. Let's define another point q where x = 4

```python
f(4)
```

```python
fig,ax = plt.subplots()
plt.axhline(y=0,color='lightgray')
plt.axvline(x=0,color='lightgray')
plt.plot(x,y)
plt.scatter(2,10,color='red',zorder=2)
plt.scatter(4,26,color='orange',zorder=2)
```

To find the slope b/w p and q point, 
m = (y2-y1)/(x2-x1)

```python
m = (26-10)/(4-2)
m
```

To plot the line passes through p and q, the equation of a straight line is y = mx + c......Then c = y - mx

```python
c = 26 - m*4
c
```

```python
line_y = m*x + c
```

```python
fig,ax = plt.subplots()
plt.axhline(y=0,color='lightgray')
plt.axvline(x=0,color='lightgray')
plt.plot(x,y)
plt.scatter(2,10,color='red',zorder=2)
plt.scatter(4,26,color='orange',zorder=2)
plt.plot(x,line_y,color='orange')
plt.xlim(-10,10)
plt.ylim(-5,150)
```
