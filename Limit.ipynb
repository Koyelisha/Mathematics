Limit

```python
def inf_fxn(my_x):
    my_y = 25/my_x
    return my_y
```

```python
inf_fxn(1e3)
```

```python
inf_fxn(1e6)
```

```python
y = inf_fxn(x)
y
```

```python
fig,ax = plt.subplots()
plt.plot(x,y)
ax.set_xlim(-10,10)
ax.set_ylim(-300,300)
```

```python
x_left = x[x<0]
x_right = x[x>0]
```

```python
left_y = inf_fxn(x_left)
right_y = inf_fxn(x_right)
```

```python
fig,ax = plt.subplots()
plt.axhline(y=0,color='lightgray')
plt.axvline(x=0,color='lightgray')
plt.plot(x_left,left_y,color='green')
plt.plot(x_right,right_y,color='green')
ax.set_xlim(-10,10)
ax.set_ylim(-300,300)
```

lim   (x^2 - 1)/(x-1)
x->0

```python
def inf_fxn(my_x):
    my_y = ((my_x**2)-1)/(my_x-1)
    return my_y
```

```python
y = inf_fxn(x)
y
```

```python
fig,ax = plt.subplots()
plt.xlim(-1,5)
plt.ylim(-1,5)
plt.axhline(y=0,color='lightgray')
plt.axvline(x=0,color='lightgray')
plt.plot(x,y)
```

sinx/x

```python
def sin_fxn(my_x):
    my_y = np.sin(my_x)/my_x
    return my_y
```

```python
my_y = sin_fxn(x)
```

```python
fig,ax = plt.subplots()
plt.axvline(x=0,color='lightgray')
plt.axhline(y=0,color='lightgray')
plt.plot(x,my_y)
plt.xlim(-7.5,7.5)
plt.ylim()
```

(x^2 - 25)/(x+5) 

```python
def fxn(my_x):
    my_y = ((my_x**2) - 25)/(x+5)
    return my_y
```

```python
my_y = fxn(x)
```

```python
fig,ax = plt.subplots()
plt.plot(x,my_y)
```
