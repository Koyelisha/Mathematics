```python
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
```

```python
x = np.linspace(-10,10,1000)
```

```python
y1 = 3 * x
```

```python
y2 = 1 + (5*x)/2
```

```python
fig,ax = plt.subplots()
plt.xlabel('x')
plt.ylabel('y')
ax.set_xlim([0,3])
ax.set_ylim([0,8])
ax.plot(x,y1)
ax.plot(x,y2)
plt.axvline(x=2,color='purple',linestyle='--')
plt.axhline(y=6,color='purple',linestyle='--')
```

```python
y3 = -5 + (2*x)/3
```

```python
y4 = (7-2*x)/5
```

```python
fig,ax = plt.subplots()
plt.xlabel('x')
plt.ylabel('y')
plt.axvline(x=0,color='lightgray')
plt.axhline(y=0,color='lightgray')
ax.set_xlim([-6,10])
ax.set_ylim([-6,10])
ax.plot(x,y3,label='y3')
ax.plot(x,y4,label='y4')
plt.axvline(x=6,color='purple',linestyle='--')
plt.axhline(y=-1,color ='purple',linestyle='--')
```

```python
y5 = 6 - x
```

```python
y6 = (16 - 2 * x)/3
```

```python
fig,ax = plt.subplots()
plt.xlabel('x')
plt.ylabel('y')
ax.set_xlim([0,6])
ax.set_ylim([0,8])
plt.plot(x,y5)
plt.plot(x,y6)
plt.axhline(y=4,color='purple',linestyle='--')
plt.axvline(x=2,color='purple',linestyle='--')
```

```python
y7 = x/4
```

```python
y8 = (2*x + 6)/5
```

```python
fig, ax = plt.subplots()
plt.xlabel('x')
plt.ylabel('y')
ax.set_xlim([-10,4])
ax.set_ylim([-4,4])
plt.axhline(y=0,color='lightgray')
plt.axvline(x=0,color='lightgray')
plt.plot(x,y7)
plt.plot(x,y8)
plt.axhline(y=-2,color='purple',linestyle='--')
plt.axvline(x=-8,color='purple',linestyle='--')
```
